```python
score=int(input("Enter your score: ")) #taking input as grade
if score>=90: #checking if the grade is above 90
  print("Grade A") #if yes it prints A
elif score>=80: #checking if the grade is in between 80-89
  print("Grade B") #if yes it prints B
elif score>=70: #checking if the grade is in between 70-79
  print("Grade C") #if yes it prints C
elif score>=60: #checking if the grade if in between 60-69
  print("Grade D") #if yes it prints D
else: #checking the grade below 60
  print("Grade F") # if yes it prints F
```

```python
students ={} #empty directory to create a key value pair
while True: #to make menu loop,it means the program keeps running until the user chooses to exit
  print("\n1. Add Student")
  print("2. Update Grade")
  print("3.View All Grades")
  print("4.Exit")
  choice = input("Enter your choice: ") #to take the input
  if choice=="1": #to add new student to directory
    name=input("Enter student name: ")
    grade=input("Enter student grade: ")
    students[name]=grade #adds new entry to directory
    print(f"{name} added successfully")
  elif choice=="2": #updates exixting student
    name=input("Enter student name: ")
    if name in students: #checking if the student name exists
      grade=input("Enter new grade: ")
      students[name]=grade #upgrading students data
      print(f"{name}'s grade updated successfully")
    else:
      print(f"{name} not found.")
  elif choice=="3": #we can view all the students and grades
    print("\nAll student grades: ")
    for name,grade in students.items():
      print(f"{name}:{grade}")
  elif choice=="4": #we exit from the loop
      print("Exiting program.")
      break
  else:
      print("Invalid choice. Please try again.")
```
